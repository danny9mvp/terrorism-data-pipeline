```python
#2003-01-01T00:00
df = crimes_raw.withColumn("date", F.to_date(F.substring('FECHA_HECHO', 0, 10), 'yyyy-MM-dd'))
```

```python
file_url = "https://miapi.com"

def batch_url_list(offset, limit):
    return [f"{file_url}$offset={i}" for i in range(offset, limit, 1000)]
```

```python
batch_list = batch_url_list(0, 40000)
print(batch_list)
```
